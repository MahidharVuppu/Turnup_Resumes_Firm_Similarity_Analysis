```python
import numpy as np
import pandas as pd
from scipy.stats import wasserstein_distance

# Function to compute the Earth Mover's Distance (EMD)
def compute_earth_movers_distance(embedding_a, embedding_b):
    # Flattening the embeddings
    firm_a_embeddings = np.array(embedding_a).flatten()
    firm_b_embeddings = np.array(embedding_b).flatten()

    # Calculate Earth Mover's Distance (EMD) using Wasserstein Distance
    emd = wasserstein_distance(firm_a_embeddings, firm_b_embeddings)

    return emd

# Extract the first 5 firms and their embeddings from the DataFrame
first_5_firms_df = df.head(5)

# Initialize an empty list to store the EMD results
results = []

# Iterate over the first 5 firms and calculate the pairwise EMD
for i in range(len(first_5_firms_df)):
    for j in range(i + 1, len(first_5_firms_df)):
        firm_a = first_5_firms_df['firm_names'].iloc[i]
        firm_b = first_5_firms_df['firm_names'].iloc[j]
        embedding_a = first_5_firms_df['embeddings'].iloc[i]
        embedding_b = first_5_firms_df['embeddings'].iloc[j]

        # Calculate the Earth Mover's Distance
        emd_distance = compute_earth_movers_distance(embedding_a, embedding_b)

        # Append the results to the list
        results.append({
            'Firm A': firm_a,
            'Firm B': firm_b,
            'EMD': emd_distance
        })

# Convert the results to a DataFrame for better readability
emd_results_df = pd.DataFrame(results)

# Display the EMD results DataFrame
print(emd_results_df)
```

```python
import numpy as np
import pandas as pd
from scipy.stats import entropy

# Function to compute centroid of a set of embeddings
def compute_centroid(embeddings):
    embeddings_array = np.array(embeddings)
    centroid = np.mean(embeddings_array, axis=0)
    return centroid

# Function to compute KL Divergence between two centroids
def compute_kl_divergence(centroid_a, centroid_b):
    centroid_a = np.clip(centroid_a, 1e-10, None)  # Avoid division by zero
    centroid_b = np.clip(centroid_b, 1e-10, None)
    centroid_a /= np.sum(centroid_a)  # Normalize to sum to 1
    centroid_b /= np.sum(centroid_b)
    kl_divergence = entropy(centroid_a, centroid_b)
    return kl_divergence

# Extract the first 5 firms and their embeddings from the DataFrame
first_5_firms_df = df.head(5)

# Initialize an empty list to store the KL divergence results
kl_results = []

# Iterate over the first 5 firms and calculate the pairwise KL Divergence
for i in range(len(first_5_firms_df)):
    for j in range(i + 1, len(first_5_firms_df)):
        firm_a = first_5_firms_df['firm_names'].iloc[i]
        firm_b = first_5_firms_df['firm_names'].iloc[j]
        embeddings_a = first_5_firms_df['embeddings'].iloc[i]
        embeddings_b = first_5_firms_df['embeddings'].iloc[j]

        # Compute centroids for both firms
        centroid_a = compute_centroid(embeddings_a)
        centroid_b = compute_centroid(embeddings_b)

        # Calculate KL Divergence
        kl_divergence = compute_kl_divergence(centroid_a, centroid_b)

        # Append the results to the list
        kl_results.append({
            'Firm A': firm_a,
            'Firm B': firm_b,
            'KL Divergence': kl_divergence
        })

# Convert the results to a DataFrame for better readability
kl_results_df = pd.DataFrame(kl_results)

# Save the DataFrame to a CSV file in the specified path
save_path = '/content/drive/MyDrive/Processed_Data/kl_divergence_results.csv'
kl_results_df.to_csv(save_path, index=False)

print(f"KL Divergence results saved to {save_path}")
```
